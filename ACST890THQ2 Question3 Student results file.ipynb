```python
## a. read data to Jupyter Notebook
import pandas as pd
dataset = pd.read_csv("~/Documents/sem1 2018/ACST890/THQ2/student_results.csv")
print(dataset)
```

```python
## b. replace NAs with 0
dataset[dataset.isnull()].head()
dataset.fillna(0)
```

```python
## c. add Total.marks as the last column with students' final mark and plot by student category
dataset['Total.marks']=100*0.07*(dataset['Assessment 1']/70)+100*0.03*(dataset['Assessment 2']/30)+100*0.07*(dataset['Assessment 3']/70)+100*0.03*(dataset['Assessment 4']/30)+100*0.03*(dataset['Online quiz 1 ']/30)+100*0.02*(dataset['Online quiz 2']/20)+100*0.03*(dataset['Online quiz 3']/30)+100*0.02*(dataset['Online quiz 4']/20)+100*0.024*(dataset['Group assignment 1']/30)+100*0.016*(dataset['Group assignment 2']/20)+100*0.04*(dataset['Group assignment 3']/50)+100*0.02*(dataset[' Group evaluation']/2)+100*0.6*(dataset['Final exam']/100)
dataset

import matplotlib.pyplot as plt
import numpy as np

dataset.boxplot(column='Total.marks',by='Student category')
```

```python
## d. calculate statistics and create data table, no class 13 and 14 as stated in question
import numpy as np
import pandas as pd
import dateutil

aa = dataset.groupby('Class No').agg({"Total.marks": [np.mean, np.var, np.std]})
np.save('stat.table',aa) # save as stat.table
aa
```

```python
## e. add Grade as a column to dataset
import pandas as pd

dataset['Grade']=pd.cut(dataset['Total.marks'], bins=[0,50,65,75,85,101],right=False,labels=['F','P','Cr','D','HD'])
dataset
```

```python
# f.  simple linear regression with Final.exam as the response and Assessment.1 as the predictor
# check final exam 

dataset['Final exam'].value_counts() #40students with mark of 10
dataset['Assessment 1'].value_counts() #52 students with 0 marks

import seaborn as sns
sns.lmplot("Assessment 1","Final exam",dataset)

from sklearn.linear_model import LinearRegression 
est = LinearRegression(fit_intercept = True)

x=dataset[['Assessment 1']]
y=dataset[['Final exam']]

est.fit(y, x)
print ("Coefficients: ", est.coef_)
print ("Intercept: ", est.intercept_)

from sklearn import metrics
yhat = est.predict(x)
print ("MSE:", metrics.mean_squared_error(yhat, y))
print ("R^2:", metrics.r2_score(yhat, y))
print ("var:", y.var())

# seems to be a postive relation between assessment 1 and final exam results. however fitting simple linear regression is not ideal.
# For both final exam and assessment 1, there are some students with low marks.
# Both response and predictor are not normally distributed. 
# Standadisation might help or seek other techniques such as generalised additive model.
# Next let's try fit more variables and check the result.
```

```python
# g. multiple regression
# subset with final exam and variables

import statsmodels.api as sm

import pandas as pd
import numpy as np
from sklearn import preprocessing
import matplotlib.pyplot as plt 
plt.rc("font", size=14)
from sklearn.linear_model import LogisticRegression
from sklearn.cross_validation import train_test_split
import seaborn as sns
sns.set(style="white")
sns.set(style="whitegrid", color_codes=True)

dataset1 = dataset[['Final exam','Assessment 1','Assessment 3','Online quiz 1 ','Group assignment 2']]
dataset1

x=dataset1[['Assessment 1','Assessment 3','Online quiz 1 ','Group assignment 2']]
y=dataset1['Final exam']

# plot y
sns.countplot(y, data=dataset1)
plt.show()

# check correlation between variables and it looks fine
sns.heatmap(dataset1.corr())
plt.show()


x = sm.add_constant(x)
est = sm.OLS(y, x).fit()

est.summary()

# The model does not perform well with a 0.16 R^2. Linear regression is not a good approach.
# However, it is obvious that assessment 1,2 and group assignment 2 are siginificant variables.
# standadisation might help perform a better result.
# Ideally we can perform a logistic regression that dealing with a big group of low values.
# Since there quite a lot of students with exam score of 10, we can fit generalised additive model as well.


```
